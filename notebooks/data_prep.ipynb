```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)

data = np.random.randn(100, 5)
df = pd.DataFrame(data=data,
                  columns=['Feature ' + str(i) for i in range(1, 6)])
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df, scaled_df.shape
```

```python
scaled_df = pd.DataFrame(data=scaled_df,
                         columns=['Feature ' + str(i) for i in range(1, 6)])
```

```python
scaled_df.to_csv('data.csv', index=False)
```
